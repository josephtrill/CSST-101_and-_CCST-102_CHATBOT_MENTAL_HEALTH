```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import datetime
import joblib

# Initialize tokenizer and models
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")

# Load pretrained models
safetensors_model1 = AutoModelForSequenceClassification.from_pretrained("model.safetensors")
safetensors_model2 = AutoModelForSequenceClassification.from_pretrained("model2.safetensors")
random_forest_model = joblib.load("random_forest_model (2).pkl")

def log_message(log_type, message):
    """Logs messages with timestamps for debugging and monitoring."""
    timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    print(f"[{timestamp}] {log_type}: {message}")

def analyze_user_input(user_input):
    """Analyzes user input using pretrained models."""
    try:
        # Tokenize the user input
        inputs = tokenizer(user_input, return_tensors="pt", truncation=True, padding=True)

        # Generate predictions from SafeTensors models
        with torch.no_grad():
            outputs1 = safetensors_model1(**inputs)
            pred1 = torch.argmax(outputs1.logits, dim=1).item()

            outputs2 = safetensors_model2(**inputs)
            pred2 = torch.argmax(outputs2.logits, dim=1).item()

        # Combine predictions (e.g., average)
        combined_preds = np.array([pred1, pred2]).mean().reshape(1, -1)

        # Generate final prediction using the Random Forest model
        rf_prediction = random_forest_model.predict(combined_preds)[0]

        return {
            "Model1 Prediction": pred1,
            "Model2 Prediction": pred2,
            "Random Forest Prediction": rf_prediction,
        }
    except Exception as e:
        log_message("Error", f"Error analyzing input: {e}")
        return None

def provide_support(predictions):
    """Provides responses and advice based on model predictions."""
    if predictions is None:
        return "I'm sorry, I couldn't analyze your input. Could you please try rephrasing?"

    advice = {
        0: "You seem to be doing well. Remember to maintain a healthy balance in your life.",
        1: "It sounds like you're facing some challenges. Consider reaching out to a friend or family member for support.",
        2: "Your situation seems serious. I strongly recommend consulting a mental health professional.",
    }

    return advice.get(predictions["Random Forest Prediction"], "I'm here to help, but I suggest seeking professional support if you're feeling overwhelmed.")

def mental_health_chatbot():
    """Main chatbot function for user interaction."""
    log_message("Info", "Chatbot session started.")
    print("Welcome to the Mental Health Support Chatbot!")
    print("Type 'exit' anytime to end the session.\n")

    while True:
        user_input = input("You: ").strip()
        if user_input.lower() == "exit":
            log_message("Info", "Chatbot session ended.")
            print("Chatbot: Take care and stay safe. Goodbye!")
            break

        predictions = analyze_user_input(user_input)
        response = provide_support(predictions)
        print(f"Chatbot: {response}")

# Run chatbot
if __name__ == "__main__":
    mental_health_chatbot()
```
