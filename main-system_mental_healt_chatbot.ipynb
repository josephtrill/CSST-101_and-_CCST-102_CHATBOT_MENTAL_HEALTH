```python
import pandas as pd
from langchain_ollama import OllamaLLM
from langchain_core.prompts import ChatPromptTemplate

# Template for the chatbot prompt
template = """
You are a helpful assistant that can answer questions based on the user's context and a CSV file's data.

Here is the conversation history:
{context}

Here is the CSV data:
{csv_data}

Question: {question}

Answer:
"""

# Initialize the model and prompt chain
model = OllamaLLM(model="qwen:0.5b")
prompt = ChatPromptTemplate.from_template(template)
chain = prompt | model

def load_csv(file_path):
    """Load a CSV file and return it as a string."""
    try:
        csv_file = pd.read_csv(file_path)
        return csv_file.to_string(index=False)  # Convert to a readable string
    except Exception as e:
        print(f"Error loading CSV: {e}")
        return None

def handle_conversation(csv_data):
    """Handle user interaction with the chatbot."""
    context = ""
    print("Welcome to the Mental Health Chatbot! Type 'exit' to quit or '/ <your question>' to query the CSV.")
    while True:
        user_input = input("You: ")
        if user_input.lower() == "exit":
            break
        if user_input.startswith("/query"):
            question = user_input.replace("/query", "").strip()
            result = chain.invoke({"context": context, "csv_data": csv_data, "question": question})
            print("Bot: ", result.strip())
            context += f"\nUser: {user_input}\nAI: {result.strip()}"
        else:
            print("Bot: Please use '/query' to ask questions about the CSV.")

if __name__ == "__main__":
    file_path = "new_data.csv"  # Replace with your CSV file path
    csv_data = load_csv(file_path)
    if csv_data:
        handle_conversation(csv_data)
```
